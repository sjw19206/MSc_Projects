```python
from bs4 import BeautifulSoup

import requests
```

```python
sefaMemberUrl = 'https://pureportal.coventry.ac.uk/en/organisations/school-of-economics-finance-and-accounting/persons/'
sefaMemberSubQuery = '?page='
sefaMemberCounter = 0
```

```python
while(True):
    finalUrl = sefaMemberUrl + sefaMemberSubQuery + str(sefaMemberCounter)
    print('Final url :: ', finalUrl)
    page = requests.get(finalUrl)
    print('page :: ', page)
    if(page.status_code == 200):
        print('Page available :: ')
        sefaMemberCounter += 1
        print('count :: ', sefaMemberCounter)
        soup = BeautifulSoup(page.text, "html.parser")  
        
        all_staff = soup.find_all("a", class_="link person")
        print(len(all_staff))
        if(len(all_staff)):
            for this_staff in all_staff:
                print(this_staff.string)    
                print(this_staff.attrs['href'])
                print("-----------------")
                
         
        else:
            break;
        
    else:
        break
```

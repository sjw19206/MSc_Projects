```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
```

```python
nltk.download("stopwords")
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download("punkt")
```

```python
stop_words = stopwords.words('english')
wnl = WordNetLemmatizer()
sb = SnowballStemmer(language='english')
```

```python
def pre_process_doc(input_document):
    # stop words
    stop_words = stopwords.words('english')
    # lemmatize
    wnl = WordNetLemmatizer()
    # stemmer
    sb = SnowballStemmer(language='english')
    
    print('<-------------Original----------------------->')
    print(input_document)
    print('<-------------Lower-case----------------------->')
    # to lower-case
    input_document = input_document.lower()
    
    print(input_document)
    print('<-------------Remove special chars----------------------->')
    
    ## remove special characters
    ##input_document = input_document.translate(str.maketrans('', '', string.punctuation))
    
    print(input_document)
    print('<------------lemm n stem------------------------>')
    
    # print('Keyword :: ', research_keywords)

    pro_inp_keywords = ' '

    input_tokens = word_tokenize(input_document)

    for input_token in input_tokens:
        input_token = input_token.translate(str.maketrans('', '', string.punctuation))
        if input_token not in stop_words and len(input_token) > 2 :    
            if not input_token.isdigit():
                lem_inp_token = wnl.lemmatize(input_token)
                # print('lem :: ', lem_res_token)
                stem_inp_token = sb.stem(lem_inp_token)
                # print('Stem ::', stem_res_token)
                pro_inp_keywords = pro_inp_keywords + stem_inp_token + ' '
    
    print(pro_inp_keywords)
    print('<--------------------------------------------->')
    # print('pro-kw :: ', pro_res_keywords)
    return pro_inp_keywords.strip()
```

```python
processed_train_doc_list = []
processed_test_doc_list = []
```

```python
for train_doc in train_news_doc_list:
    processed_train_doc = pre_process_doc(train_doc)
    processed_train_doc_list.append(processed_train_doc)
```

```python
for test_doc in test_news_doc_list:
    processed_test_doc = pre_process_doc(test_doc)
    processed_test_doc_list.append(processed_test_doc)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
train_vector = vectorizer.fit_transform(processed_train_doc_list)
```

```python
feature_names = vectorizer.get_feature_names_out()
print(feature_names)
print(len(feature_names))
```

```python
from sklearn.cluster import KMeans
val_num_clusters = 5
val_max_iter = 100
val_tol = 0.0002
model = KMeans(n_clusters = val_num_clusters, max_iter = val_max_iter, tol = val_tol)
model.fit(train_vector)
```

```python
print("cluster no. of input documents, in the order they received:")
print(model.labels_)
```

```python
# the clusters generated by k-means.
common_words = model.cluster_centers_.argsort()[:,-1:-26:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(feature_names[word] for word in centroid))
```

```python
for test_doc in processed_test_doc_list:
    test_doc_vector = vectorizer.transform([test_doc])
    prediction = model.predict(test_doc_vector)
    print('<----------------------------------------->')
    print('doc :: ', test_doc, "- cluster :: ", prediction)
```
